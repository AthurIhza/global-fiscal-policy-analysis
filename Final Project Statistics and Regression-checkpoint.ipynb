```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns

# =====================================================================
# INITIALIZATION & DATA PREPARATION
# =====================================================================
# Load global dataset (Make sure the file name matches your local directory)
df_global = pd.read_csv('world_bank_data_2025.csv')

# Feature Engineering Category A: Global Crisis Eras
def determine_era(year):
    if year < 2020: 
        return 'Pre_COVID'
    elif year in [2020, 2021]: 
        return 'COVID_Crisis'
    else: 
        return 'Post_COVID'

df_global['Era'] = pd.Categorical(
    df_global['year'].apply(determine_era), 
    categories=['Pre_COVID', 'COVID_Crisis', 'Post_COVID'], 
    ordered=True
)

# Feature Engineering Category B: Global Income Clusters
df_global['Income_Group'] = pd.qcut(
    df_global['GDP per Capita (Current USD)'], 
    q=3, 
    labels=['Low_Income', 'Medium_Income', 'High_Income']
)

# Missing Value Cleaning for Selected Features
num_features = ['Government Expense (% of GDP)', 'Tax Revenue (% of GDP)', 'Public Debt (% of GDP)']
cat_features = ['Era', 'Income_Group']
target = 'GDP Growth (% Annual)'
df_clean = df_global.dropna(subset=num_features + cat_features + [target]).copy()

# One-Hot Encoding (drop_first=True fixes the reference baseline control)
df_model = pd.get_dummies(df_clean, columns=cat_features, drop_first=True)
dummy_cols = ['Era_COVID_Crisis', 'Era_Post_COVID', 'Income_Group_Medium_Income', 'Income_Group_High_Income']
for col in dummy_cols:
    df_model[col] = df_model[col].astype(int)

# Feature Engineering: Interaction (Moderation) Term
df_model['GovExpense_x_COVID'] = df_model['Government Expense (% of GDP)'] * df_model['Era_COVID_Crisis']

# Feature Engineering: Log Transformation (Mitigating Public Debt Skewness)
df_model['Log_Public_Debt'] = np.log1p(df_model['Public Debt (% of GDP)'])


# =====================================================================
# OBJECTIVE 1: Descriptive Statistics (Global Baseline Rates)
# =====================================================================
print("\n" + "="*80)
print("OBJECTIVE 1: DESCRIPTIVE STATISTICS (Global Baseline Rates)")
print("="*80)
print(df_clean[[target] + num_features].describe().T)


# =====================================================================
# OBJECTIVE 2: Mean Differences using Independent Sample T-Test
# =====================================================================
print("\n" + "="*80)
print("OBJECTIVE 2: INDEPENDENT SAMPLE T-TEST (High Income vs Low Income)")
print("="*80)
group_high = df_clean[df_clean['Income_Group'] == 'High_Income'][target]
group_low = df_clean[df_clean['Income_Group'] == 'Low_Income'][target]

t_stat, p_val = stats.ttest_ind(group_high, group_low, nan_policy='omit')
print(f"Mean GDP Growth High Income : {group_high.mean():.2f}%")
print(f"Mean GDP Growth Low Income  : {group_low.mean():.2f}%")
print(f"T-Statistic                  : {t_stat:.4f}")
print(f"P-Value                      : {p_val:.4f}")
if p_val < 0.05:
    print("Conclusion: The difference in mean economic growth is statistically SIGNIFICANT.")
else:
    print("Conclusion: The difference in mean economic growth is NOT statistically significant.")


# =====================================================================
# OBJECTIVE 3: Simple Regression Model (Baseline)
# =====================================================================
print("\n" + "="*80)
print("OBJECTIVE 3: SIMPLE REGRESSION MODEL (Baseline)")
print("="*80)
print("Theoretical Equation:")
print("GDP_Growth = β0 + β1(Gov_Expense) + ε")
print("-"*80)

X_obj3 = sm.add_constant(df_model[['Government Expense (% of GDP)']])
model_simple = sm.OLS(df_model[target], X_obj3).fit()
print(f"Estimated Results: GDP_Growth = {model_simple.params['const']:.4f} + {model_simple.params['Government Expense (% of GDP)']:.4f}*(Gov_Expense)")
print(f"R-squared : {model_simple.rsquared:.4f}")


# =====================================================================
# OBJECTIVE 4: Multiple Regression Model (Controlled)
# =====================================================================
print("\n" + "="*80)
print("OBJECTIVE 4: MULTIPLE REGRESSION MODEL (Controlled)")
print("="*80)
print("Theoretical Equation:")
print("GDP_Growth = β0 + β1(Gov_Expense) + β2(Tax_Revenue) + β3(Public_Debt)")
print("             + β4(Era_COVID_Crisis) + β5(Era_Post_COVID)")
print("             + β6(Income_Group_Medium_Income) + β7(Income_Group_High_Income) + ε")
print("-"*80)

X_obj4 = sm.add_constant(df_model[num_features + dummy_cols])
model_controlled = sm.OLS(df_model[target], X_obj4).fit()
print(f"R-squared : {model_controlled.rsquared:.4f} (Increase vs Baseline: {model_controlled.rsquared - model_simple.rsquared:.4f})")


# =====================================================================
# OBJECTIVE 5: Interaction Effects (Moderation Model)
# =====================================================================
print("\n" + "="*80)
print("OBJECTIVE 5: INTERACTION MODEL (Moderation Analysis)")
print("="*80)
print("Theoretical Equation:")
print("GDP_Growth = β0 + β1(Gov_Expense) + β2(Tax_Revenue) + β3(Public_Debt)")
print("             + β4(Era_COVID_Crisis) + β5(Era_Post_COVID)")
print("             + β6(Income_Group_Medium_Income) + β7(Income_Group_High_Income)")
print("             + β8(Gov_Expense × Era_COVID_Crisis) + ε")
print("-"*80)

features_obj5 = num_features + dummy_cols + ['GovExpense_x_COVID']
X_obj5 = sm.add_constant(df_model[features_obj5])
model_interaction = sm.OLS(df_model[target], X_obj5).fit()
print(f"R-squared : {model_interaction.rsquared:.4f}")
print(f"Interaction P-Value (GovExpense x COVID): {model_interaction.pvalues['GovExpense_x_COVID']:.4f}")


# =====================================================================
# OBJECTIVE 6: Data Transformation Comparison (Log-Linear Model)
# =====================================================================
print("\n" + "="*80)
print("OBJECTIVE 6: DATA TRANSFORMATION COMPARISON (Log-Linear)")
print("="*80)
print("Theoretical Equation (With Log Transformation on Public Debt):")
print("GDP_Growth = β0 + β1(Gov_Expense) + β2(Tax_Revenue) + β3(LN_Public_Debt)")
print("             + β4(Era_COVID_Crisis) + β5(Era_Post_COVID)")
print("             + β6(Income_Group_Medium_Income) + β7(Income_Group_High_Income)")
print("             + β8(Gov_Expense × Era_COVID_Crisis) + ε")
print("-"*80)

features_obj6 = ['Government Expense (% of GDP)', 'Tax Revenue (% of GDP)', 'Log_Public_Debt'] + dummy_cols + ['GovExpense_x_COVID']
X_obj6 = sm.add_constant(df_model[features_obj6])
model_transformed = sm.OLS(df_model[target], X_obj6).fit()

print("MODEL ACCURACY COMPARISON:")
print(f"R-squared Without Transformation (Obj 5) : {model_interaction.rsquared:.4f}")
print(f"R-squared With Transformation (Obj 6)    : {model_transformed.rsquared:.4f}")

print("\n" + " FINAL TRANSFORMED MODEL SUMMARY (OBJECTIVE 6) ".center(80, "#"))
print(model_transformed.summary())
```

```python
# =====================================================================
# DIAGNOSTIC VISUALIZATIONS FOR PORTFOLIO REPORT
# =====================================================================
sns.set_theme(style="whitegrid")

# Chart 1: R-squared Progression Tracking Across the 4 Models
plt.figure(figsize=(10, 5))
stages = ['Obj 3\nSimple Baseline', 'Obj 4\nMultiple Controlled', 'Obj 5\nInteraction Model', 'Obj 6\nLog Transformed']
r_values = [model_simple.rsquared, model_controlled.rsquared, model_interaction.rsquared, model_transformed.rsquared]
ax = sns.barplot(x=stages, y=r_values, palette='Blues_d')

for p in ax.patches:
    ax.annotate(f"{p.get_height():.4f}", (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 9), textcoords='offset points', fontweight='bold', fontsize=10)

plt.title('Model Explanatory Power Comparison (R-squared Tracking)', fontsize=12, fontweight='bold', pad=15)
plt.ylabel('R-squared Value')
plt.ylim(0, max(r_values) * 1.2)
plt.tight_layout()
plt.show()
```

```python
# Chart 2: Final Model Coefficients Margins Plot (Objective 6)
coef_df = pd.DataFrame({
    'Feature': model_transformed.params.index,
    'Coefficient': model_transformed.params.values,
    'Lower_CI': model_transformed.conf_int()[0].values,
    'Upper_CI': model_transformed.conf_int()[1].values
}).query("Feature != 'const'").sort_values(by='Coefficient')

plt.figure(figsize=(10, 6))
plt.errorbar(x=coef_df['Coefficient'], y=coef_df['Feature'], 
             xerr=[coef_df['Coefficient'] - coef_df['Lower_CI'], coef_df['Upper_CI'] - coef_df['Coefficient']],
             fmt='o', color='darkblue', ecolor='red', elinewidth=2, capsize=5, markersize=8)
plt.axvline(x=0, color='black', linestyle='--', linewidth=1.2)
plt.title('Coefficient Plot with 95% Confidence Interval (Final Transformed Model)', fontsize=12, fontweight='bold', pad=15)
plt.xlabel('Estimated Marginal Impact on GDP Growth')
plt.tight_layout()
plt.show()
```
